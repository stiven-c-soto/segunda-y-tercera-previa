```python
#se convierte la cadena que almacena octal en una lista de strings ( " " )
class octal:
  lista_octal = []     #declaramos la variable lista_octal como una lista

  def __init__(self, cadena_octal):
    self.cadena_octal = cadena_octal
    self.lista_octal = list( self.cadena_octal )    #hacemos que self.lista_octal tenga de contenido la cadena pero convertida en lista
  
  def string( self ):
    print(self.cadena_octal)
    print(self.lista_octal)

a = octal( "2345" )
a.string()
```

```python
class octal:
  lista_octal = []

  def __init__(self, cadena_octal):
    self.cadena_octal = cadena_octal
    self.lista_octal = list( self.cadena_octal )
    i = 0
    for dato in self.lista_octal :      # con este for volvemos el contenido de self.lista_octal en enteos, en un numero
      self.lista_octal[i] = int(dato)
      i = i + 1

    self.lista_octal = list(reversed(self.lista_octal))   # para que la cadena concuerde con el orden en el que se opera 
                                                          # un dato octal la revertimos con esa funcion
  
  def string( self ):
    print(self.cadena_octal)
    print(self.lista_octal)

a = octal( "2345" )
a.string()
```
